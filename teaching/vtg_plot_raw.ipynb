```python
import csv
import matplotlib.pyplot as plt

# Arrays of data

# IMU
imutime = []
ax = []
ay = []
az = []
gx = []
gy = []
gz = []
mx = []
my = []
mz = []

# GPS
gpstime = []
lat = []
lon = []
alt = []
```

```python
# parse csv data

with open('vtg_log0.csv', 'r') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        if row[1] == '2':
            imutime.append(float(row[0]) / 1000.0)
            ax.append(row[2])
            ay.append(row[3])
            az.append(row[4])
            gx.append(row[5])
            gy.append(row[6])
            gz.append(row[7])
            mx.append(row[8])
            my.append(row[9])
            mz.append(row[10])
        elif row[1] == '1':
            gpstime.append(float(row[0]) / 1000.0)
            lon.append(row[2])
            lat.append(row[3])
            alt.append(row[4])
```

```python
# plot things

plt.plot(imutime, ax, 'r', imutime, ay, 'b', imutime, az, 'g')
plt.xlabel("Time (s)")
plt.ylabel("Acceleration (m/s^2)")
plt.legend(['x', 'y', 'z'])
plt.show()
```
